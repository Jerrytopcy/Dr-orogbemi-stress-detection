```

import pandas as pd
import hashlib

# Simulate a user database
users_db = pd.DataFrame(columns=['name', 'email', 'password_hash'])

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

def register_user(name, email, password):
    if email in users_db['email'].values:
        return "Email already registered."
    users_db.loc[len(users_db)] = [name, email, hash_password(password)]
    return "User registered successfully."

def login_user(email, password):
    hashed = hash_password(password)
    user = users_db[users_db['email'] == email]
    if not user.empty and user.iloc[0]['password_hash'] == hashed:
        return f"Welcome, {user.iloc[0]['name']}!"
    return "Invalid email or password."
```

```

from sklearn.ensemble import RandomForestClassifier
import numpy as np
import joblib

# Simulated training data (for demo purpose only)
X_train = np.random.randint(0, 5, (100, 15))  # 15 questions from PSS
y_train = np.random.choice(['Low', 'Moderate', 'High'], size=100)

model = RandomForestClassifier()
model.fit(X_train, y_train)

# Save the model
joblib.dump(model, 'stress_model.pkl')
```

```

def predict_stress(user_answers):
    model = joblib.load('stress_model.pkl')
    prediction = model.predict([user_answers])[0]
    return prediction
```

```

import datetime

assessment_history = pd.DataFrame(columns=['email', 'timestamp', 'answers', 'stress_level'])

def save_assessment(email, answers, level):
    global assessment_history
    assessment_history.loc[len(assessment_history)] = [
        email,
        datetime.datetime.now(),
        answers,
        level
    ]
    return "Assessment saved."
```

```

def get_user_history(email):
    return assessment_history[assessment_history['email'] == email]
```

```

import matplotlib.pyplot as plt

def plot_stress_trend(email):
    history = get_user_history(email)
    if history.empty:
        print("No assessment history.")
        return

    plt.figure(figsize=(10, 5))
    plt.plot(history['timestamp'], history['stress_level'], marker='o')
    plt.title('Stress Trend Over Time')
    plt.xlabel('Date')
    plt.ylabel('Stress Level')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```

print(register_user("Rahmat Titilayo", "rahmat@example.com", "securepass123"))
print(login_user("rahmat@example.com", "securepass123"))

answers = np.random.randint(0, 5, size=15).tolist()
stress_level = predict_stress(answers)
print("Predicted Stress Level:", stress_level)

save_assessment("rahmat@example.com", answers, stress_level)
plot_stress_trend("rahmat@example.com")
```
